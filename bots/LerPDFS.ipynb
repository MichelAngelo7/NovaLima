```python
import PyPDF2
import json
import time
import requests
import io
import re
```

```python

with open('linkParaAtas.json', 'r') as atas:
    dados = json.load(atas)
    
    
#pegar o primeiro valor do dictionary         
data = dados.get('0')
atas = []

for dado in data: 
    url = f'{dado}'
    filtrar = f'"{dado}"'
    resposta = requests.get(url)
    try:
        with io.BytesIO(resposta.content) as abrir_pdf:
            ler_pdf = PyPDF2.PdfFileReader(abrir_pdf)
            #print(filtrar)
            filtro = re.findall('ata(.*)\"', filtrar)
            print(f"Ata: {filtro} NumeroPaginas: {ler_pdf.numPages}")
            atas.append(f"Ata: {filtro} NumeroPaginas: {ler_pdf.numPages}")
            time.sleep(10)
            abrir_pdf.close()
    except:
        print("Erro desconhecido")
        pass
        
dic_atas = {}
dic_atas[0] = atas

atas_json = json.dumps(dic_atas)
arquivo = open('PaginasParaAtas.json', 'w')
arquivo.write(atas_json)
arquivo.close()
    
```
